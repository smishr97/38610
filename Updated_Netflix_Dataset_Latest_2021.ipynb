```python
import os
import urllib.request
from tqdm.notebook import tqdm_notebook as tqdm
import requests

new_dir_path = '/content/drive/My Drive/38610/datasets/Netflix'
try:
    os.mkdir(new_dir_path)
except:
    print('Directory already exists!')

df['Local Images'] = None
for i in tqdm(range(df.shape[0])):
    # urllib.request.urlretrieve(df['Image'][i], f"{new_dir_path}/image_{i}")
    url = df['Image'][i]
    if url == url:
        response = requests.get(url)
        if response.status_code == 200:
            with open(f"{new_dir_path}/image_{i}", 'wb') as f:
                f.write(response.content)
            df['Local Images'][i] = f"{new_dir_path}/image_{i}"
```

3. Add a column to the table named `IMDb Availability`. If the link `IMDb Link` is unavailable, then populate the record as False. True otherwise.

```python
tqdm.pandas()

df['IMDb Availability'] = False

def check_link ( url ):
    if url == url:
        response = requests.get(url)
        if response.status_code == 200:
            return True
    else: 
        return False

df['IMDb Availability'] = df['IMDb Link'].progress_apply(lambda x: check_link(x))
```

4. Make a histogram using each unique genre in the `Genre` column. 

**Hint**: split values in `Genre` column if needed.

```python
unique_genre = df['Genre'].str.strip().str.split(',')
genre = unique_genre.explode().to_list()
genre = [str(x) for x in genre]
genre = [x.strip() for x in genre]

counter = {x:genre.count(x) for x in genre}
```

```python
counter
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(25,10))
plt.bar(counter.keys(), counter.values(), color='r')
# plt.hist(counter)
plt.show()
```
